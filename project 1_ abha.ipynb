# Comparing methods to elicit truthful information on sensitive behaviors among young Indian males

The purpose of this study is to determine the relative effectiveness of two established methods of asking questions in the survey for eliciting information about the prevalence of sensitive behaviors. Due to social desirability bias, survey respondents generally underreport socially undesirable or sensitive behaviors and overreport socially desirable ones. 

Survey questions asking about taboo topics such as sexual violence, domestic violence, and homosexuality often generate inaccurate survey estimates. This study will contribute towards developing and testing a methodology to generate unbiased prevalence estimates of homosexual identity, homosexual experience, sexual molestation of females, and perception towards violence against women, among Indian males.

The study followed an experimental protocol that entails collection of data through online surveys of a sample of 712 college students in India. The subjects in the experiment were assigned to treatment and control groups of four different treatment arms by a random draw.
