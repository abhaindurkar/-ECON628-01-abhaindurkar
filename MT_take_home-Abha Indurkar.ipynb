### 5.1 Calculate yearly liquor sales for each store in 2015.

### 5.2 Calculate the profit each store is making in 2015.


```python
Iowa2015fresh = Iowa2015fresh.rename(index=str, columns={ "Sale (Dollars)": "Sale_Dollars", "State Bottle Cost": "State_Bottle_Cost", "State Bottle Retail": "State_Bottle_Retail", "Bottles Sold": "Bottles_Sold"})
```

```python
Iowa2015fresh['Month_dummy']= Iowa2015fresh['salesdummy']
Iowa2015fresh.loc[Iowa2015fresh.Month_dummy == 12, 'Month_dummy']=1
Iowa2015fresh
```

```python
yearlysale=Iowa2015fresh.groupby(['Store Number','Month_dummy'])['Sale_Dollars'].sum().reset_index()
yearlysale.head()
```

```python
Iowa2015fresh['Cost']=(Iowa2015fresh.Bottles_Sold) * (Iowa2015fresh.State_Bottle_Cost)
yearlycost=Iowa2015fresh.groupby(['Store Number','Month_dummy'])['Cost'].sum().reset_index()
yearlycost.head(5)
```

```python
Iowa2015profit=pd.merge(yearlysale, yearlycost, on=['Store Number','Month_dummy'], how='left')
Iowa2015profit.head(5)
```

```python
Iowa2015profit['Profit']=Iowa2015profit.Sale_Dollars - Iowa2015profit.Cost
Iowa2015profit.head(5)
```

These are the yearly profit incurred by each store in 2015 (This calculation however, assumes that the selling price of each bottle is equal to the dollar sale which is according to the website the amount store paid, and cost price of each bottle is the cost paid by alchoholic breverages devision for each bottle multiplied by the number of bottles, this assumption is required as the data according to the website has no data of information on how many bottles the store sells, it only gives how many bottles the store ordered)
